```python
# import libraries
import pandas as pd
import numpy as np

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv')
```

```python
# check head
df.head()
```

```python
# check info
df.info()
```

### Create a basic linechart

```python
# transfrom datatype with pd.to_datetime
df['Month'] = pd.to_datetime(df['Month'])
```

```python
# check info
df.info()
```

```python
# example basic linechart
plt.figure(figsize = (20,5))
plt.plot(df['Month'] # X
        ,df['Passengers'] # Y
        ,label = 'Passenger Count' # label
        ,color = 'r' # color
        ,marker = 'v' # shape
        ,linestyle = '-' # line
        )
plt.title('International Airline Passengers Overtime') # chart title
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation = 45
          ,ha = 'center'
          )
plt.grid(linestyle = '--'
        ,alpha = 0.8 # opacity
        )
plt.show
```

### Bar chart

```python
# load dataset
tips = sns.load_dataset('tips')
```

```python
# example of vertical bar chart
plt.figure(figsize = (7,4))
plt.bar(tips['day'] # X
       ,tips['tip']) # Y
plt.title('Total Tips by Day')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.grid(axis='y', linestyle = '--', alpha = 0.6)
plt.show()
```

```python
# show aggregation sum
tips.groupby('day')['tip'].sum()
```

```python
# example of vertical bar chart using seaborn
plt.figure(figsize = (7,4))
sns.barplot(data = tips
           ,x = 'day'
           ,y = 'tip'
           ,palette = 'Blues'
           )
plt.title('Total Tips by Day')
plt.xlabel('Day')
plt.ylabel('Total Tips')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
plt.show()
```

```python
# example of histogram using matplotlib
plt.figure(figsize = (7,4))
plt.hist(tips['total_bill']
        ,bins = 10
        ,color = 'b'
        ,alpha = 0.5
        ,edgecolor = 'r'
        )
plt.title('Distribution of Total Bill Amounts')
plt.xlabel('Total Bill Amount')
plt.ylabel('Frequency')
plt.grid(axis = 'y', linestyle = '--', alpha = 0.6)
plt.show()
```

```python
# example of scatter plot using matplotlb
plt.figure(figsize = (8,4))
plt.scatter(tips['total_bill']
           ,tips['tip']
           ,color='green'
           ,marker='o'
           ,s = 20
           ,alpha = 0.7
           ,edgecolors = 'black'
           ,linewidths = 1)
plt.title('Scatter Plot if Total bill vs. Tip (Matplotlib)')
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```

```python
# create a correlation matriz
tips_num = tips.select_dtypes(['int64', 'float64'])
corr = tips_num.corr()
```

```python
# customizing the seaborn heat map appearance
plt.figure(figsize = (5,5))
sns.heatmap(corr,
           annot = True,
           fmt = '.2f',
           cbar_kws = {'label' : 'Correlation Coefficient'})
plt.title('Customized Heat Map of Correlations (Seaborn)')
plt.show()
```

```python
# calculate the sum of total bills for different days
bill_sum_by_day = tips.groupby('day')['total_bill'].sum()

# labels for each slice
labels = bill_sum_by_day.index
bill_sum_by_day
```

```python
# display percentage and absolute values
plt.figure(figsize = (5,5))
plt.pie(bill_sum_by_day,
       labels = labels,
       autopct = '%1.1f%%',
       startangle = 90,
       pctdistance = 0.5,
       explode = (0,0,0,0.1))
plt.title('Pie Chart with Percentage and Absolute Values (Seaborn)')
plt.legend(labels, loc = 'best', bbox_to_anchor = (1.0,1.0))
plt.show()
```
